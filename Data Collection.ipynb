# Web Scraping

```python
import requests
from bs4 import BeautifulSoup
```

```python
import requests
from bs4 import BeautifulSoup
url = "https://en.wikipedia.org/wiki/List_of_Indian_Premier_League_seasons_and_results"
headers = {
    "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) "
                  "AppleWebKit/537.36 (KHTML, like Gecko) "
                  "Chrome/123.0.0.0 Safari/537.36",
    "Accept-Language": "en-US,en;q=0.9",
}
page = requests.get(url, headers=headers)
page.raise_for_status()   # will throw if blocked
soup = BeautifulSoup(page.text, "html.parser")
print(soup.prettify()[:1000])
```

```python
print(soup.prettify()[:1000])
```

```python
print(soup.find('table'))
```

```python
table = pd.read_html(page.text)
ipl = table[2]
ipl.head(20)
```

# APIs

```python
from io import StringIO
api = "https://en.wikipedia.org/api/rest_v1/page/html/List_of_Indian_Premier_League_seasons_and_results"
header={
    "User-Agent": "Mozilla/5.0 (compatible; IPL-Data-Scraper/1.0; contact@example.com)"
}
response = requests.get(api, headers=header)
response.raise_for_status()
html = response.text
tables = pd.read_html(StringIO(html))
ipl = tables[4]
ipl.head(10)
```

# Image

```python
import requests
from PIL import Image
from io import BytesIO
img_url = "https://upload.wikimedia.org/wikipedia/commons/f/f7/Kohli_ruling_despite_RCB_collapse.jpg"
headers = {
    "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) "
                  "AppleWebKit/537.36 (KHTML, like Gecko) "
                  "Chrome/123.0.0.0 Safari/537.36"
}
resp = requests.get(img_url, headers=headers, timeout=15)
resp.raise_for_status()
img = Image.open(BytesIO(resp.content))
img.show()
```
