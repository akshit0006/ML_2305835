```python
x1=[3,4,5,6,2]
x2=[8,5,7,3,1]
y=[-3.7,3.5,2.5,11.5,5.7]
sx1=0; sx2=0; sy=0
ssqx1=0; ssqx2=0; ssqy=0
sx1y=0; sx2y=0; sx1x2=0
for i in range(len(x1)):
    sx1+= x1[i]; ssqx1+=x1[i]**2
    sx2+= x2[i]; ssqx2+=x2[i]**2
    sy+= y[i]; ssqy+=y[i]**2
    sx1y+= x1[i]*y[i]; sx2y+=x2[i]*y[i]; sx1x2+= x1[i]*x2[i]
mx1= sx1/len(x1)
mx2=sx2/len(x2)
my=sy/len(y)
B1= ((ssqx2 * sx1y) - (sx1x2 * sx2y))/((ssqx1 * ssqx2) - (sx1x2)**2)
B2= ((ssqx1 * sx2y) - (sx1x2 * sx1y))/((ssqx1 * ssqx2) - (sx1x2)**2)
B0 = my - (B1*mx1) - (B2*mx2)
print("VALUES OF B1, B2, B0:", B1, B2,B0)
xx1 = eval(input("ENTER THE VALUE OF X1"))
xx2 = eval(input("ENTER THE VALUE OF X2"))
yy= B0 + B1*xx1 + B2*xx2
print("VALUE OF Y=",yy)
```
