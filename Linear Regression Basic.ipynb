```python
def linreg(x,y):
    xmean = sum(x)/len(x)
    ymean= sum(y)/len(y)
    varx = 0
    cov = 0
    for i in x:
        varx = varx + (i-xmean)**2
    for i in range(len(y)):
        cov +=(x[i]-xmean)*(y[i]-ymean)
    varx = varx/len(x)
    cov = cov/len(x)
    m= cov/varx
    b= ymean-m*xmean
    return m,b

def predict(x,m,b):
    y = m*x + b
    return y
def findError(x,y,m,b):
    MSE=0
    for i in range(len(x)):
        MSE = MSE+(y[i] - predict(x[i],m,b))**2
    MSE=MSE/len(x)
    print("ERROR= ", MSE)
x = [1,2,3,4,5]
y=[10,12,15,17,18]
m,b = linreg(x,y)
print("Slope = ", m)
print("Intercept = ",b)
findError(x,y,m,b)

x = eval(input("Enter value of x for prediction:"))
print("Predicted Value of y: ", predict(x,m,b))
```
